```python
import pandas as pd
```

```python
dataf = pd.read_csv("/kaggle/input/data-jobs-by-lukebarousse/data_jobs.csv")

dataf.head()
```

```python
dataf.iloc[:5, 2:6]
```

```python
dataf.dropna(subset = "salary_hour_avg").loc[:500, ["job_title_short", "job_country", "salary_hour_avg"]]
```

```python
dataf.head()
```

```python
#copying dataframe
dataf_copied = dataf.copy()

dataf_copied.head(3)
```

```python
id(dataf) == id(dataf_copied)
```

```python
#median values
median_salary_year_avg = dataf_copied.salary_year_avg.median()

median_salary_hour_avg = dataf_copied.salary_hour_avg.median()
```

```python
#fill NA vlaues of salary_year_avg and salary_hour_avg with their respective median values

dataf_copied["salary_year_avg"] = dataf_copied.salary_year_avg.fillna(median_salary_year_avg)

dataf_copied["salary_hour_avg"] = dataf_copied.salary_hour_avg.fillna(median_salary_hour_avg)
```

```python
dataf_copied.loc[2001:2008, ["salary_year_avg", "salary_hour_avg"]]
```

```python
dataf_copied.head(2)
```

```python
#drop duplicate columns
dataf_copied = dataf_copied.drop_duplicates(subset = ["job_title", "company_name"])

dataf_copied
```
