```python
import pandas as pd
```

```python
dataf = pd.read_csv("/kaggle/input/data-jobs-by-lukebarousse/data_jobs.csv")

dataf.head(3)
```

```python
country = dataf[dataf["job_country"] == "India"]

country.head()
```

```python
country.dropna(subset = ["salary_year_avg"], inplace = True)

country
```

```python
country.info()
```

```python
country.groupby("job_title_short")["salary_year_avg"].mean().sort_values(ascending = False)
```

```python
country.groupby("job_title_short")["salary_year_avg"].agg(["min", "max"])
```

```python
country.groupby("job_title_short").size().sort_values(ascending = False)
```

```python
country["job_title_short"].value_counts()
```

```python
country.groupby("job_title_short")["salary_year_avg"].agg(["min", "max", "median", "count"]).sort_values(by = "median", ascending = False)
```
